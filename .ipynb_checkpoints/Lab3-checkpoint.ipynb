```python
class Graph:

    # using hash map
    def __init__(self, edges):
        self.graph = {}
        for x,y in edges:
            if x not in self.graph.keys():
                self.graph[x]=[]
            self.graph[x].append(y)
            # undirected
            if y not in self.graph.keys():
                self.graph[y]=[]
            self.graph[y].append(x)

    def has_edge(self, src, dst):
        value = True if src in self.graph[dst] or dst in self.graph[src] else False
        return value
    
    def add_edge(self,node1, node2):
        if node1 not in self.graph.keys():
            self.graph[node1]=[]
        self.graph[node1].append(node2)
        if node2 not in self.graph.keys():
            self.graph[node2]=[]
        self.graph[node2].append(node1)

    def get_edges(self, node):
        return self.graph[node]

    def get_graph_size(self,):
        return len(self.graph)
    
    def get_nodes(self,):
        return self.graph.keys()
    
    def get_graph(self,):
        return self.graph
    
    # Part 1.3
    def has_cycle(self,):
        seen = set()
        onpath = [False] * self.get_graph_size()
        pathdepth = [0] * self.get_graph_size()
        for node in self.graph.keys():
            if node not in seen:
                if has_cycle_each(self, node, seen, onpath, pathdepth, 0):
                    return True
        return False
    
    # Part 1.4
    def is_connected(self,):
        seen = set()
        id = [-1] * self.get_graph_size()
        count = 0
        for node in self.graph.keys():
            if node not in seen:
                dfs_count(self, node, seen, id, count)
                count = count + 1
        for i in self.graph.keys():
            if id[i] != 0:
                return False
        return True
    
    # Part 1.5
    def add_node(self, node):
        if node not in self.graph.keys():
            self.graph[node] = []
            
    def remove_edge(self, node):
        self.graph[node] = []
        for i in self.graph.keys():
            if node in self.graph[i]:
                self.graph[i] = [n for n in self.graph[i] if n!= node]
                
    def get_all_edges(self):
        all_edges = []
        for key in self.graph.keys():
            edge = self.graph[key]
            for e in edge:
                all_edges.append([key, e])
        return all_edges
            
```

```python

def BFS2(G, n1, n2):
    stack = [[n1]]
    graph = G.get_graph()
    seen = set()
    seen.add(n1)

    while len(stack) > 0:
        path = stack[0]
        node = path[-1]
        stack = stack[1:]
        for nn in graph[node]:
            if nn not in seen:
                new_path = list(path)
                new_path.append(nn)
                stack.append(new_path)
                seen.add(nn)
                if nn == n2:
                    return new_path

#g1 = Graph([[0,1],[1,2],[2,0],[3,2],[4,3],[5,4],[6,5]])
#print(BFS2(g1, 0, 6))

def DFS2(G, n1, n2):
    stack = [[n1]]
    graph = G.get_graph()
    seen = set()
    seen.add(n1)

    while len(stack) !=0:
        path = stack.pop()
        node = path[-1]
        for nn in graph[node]:
            if nn not in seen: 
                new_path = list(path)
                new_path.append(nn)
                stack.append(new_path)
                seen.add(nn)
                if nn == n2:
                    return new_path
#print(DFS2(g1, 0, 6))                
```

```python
def BFS3(G, node):
    stack = [node]
    graph = G.get_graph()
    seen = set()
    predecessor = {}
    seen.add(node)

    while len(stack) !=0:
        n1 = stack[0]
        stack = stack[1:]
        for nn in graph[n1]:
            if nn not in seen: 
                stack.append(nn)
                seen.add(nn)
                predecessor[nn] = n1
    return predecessor
#print(BFS3(g1,0))
def DFS3(G, node):
    stack = [node]
    graph = G.get_graph()
    seen = set()
    predecessor = {}
    seen.add(node)

    while len(stack) !=0:
        n1 = stack.pop()
        for nn in graph[n1]:
            if nn not in seen: 
                stack.append(nn)
                seen.add(nn)
                predecessor[nn] = n1
    return predecessor
#print(DFS3(g1,0))
```

### Part 1.3

```python
def has_cycle_each(G, node, seen, onpath, pathdepth, depth):
    graph = G.get_graph()
    depth = depth + 1
    seen.add(node)
    onpath[node] = True
    pathdepth[node] = depth

    for nn in graph[node]:
        if onpath[nn]:
            if pathdepth[nn] < depth - 1:
                return True
        if nn not in seen:
            result = has_cycle_each(G, nn, seen, onpath, pathdepth, depth)
            if result == True:
                return True
    onpath[node] = False
    return False

#print(g1.has_cycle())
```

```python
def dfs_count(G, node, seen, id, count):
    graph = G.get_graph()
    seen.add(node)
    id[node] = count
    for nn in graph[node]:
        if nn not in seen:
            dfs_count(G,nn,seen,id,count)
g2 = Graph([[0,1],[1,2],[2,3]])
#print(g2.is_connected())
    
```

```python
def has_cycle_exp(edge):
    runs = 1000
    count = 0
    for _ in range(runs):
        my_graph = create_random_graph(100,edge)
        if my_graph.has_cycle():
            count += 1
    plt.pie([count, runs - count], labels=["Has A Cycle", "No Cycle"], autopct='%1.1f%%', startangle=140)
    plt.axis('equal')
    plt.title('100 nodes, ' + str(edge) + 'edges')
    plt.show()
has_cycle_exp(25)
has_cycle_exp(50)
has_cycle_exp(60)
has_cycle_exp(75)
has_cycle_exp(100)
```

```python
def vertex(graph):
    v = []
    for node in graph.get_nodes():
        if len(graph.get_edges(node)) >= 1:
            v.append(node)
    
    return v

def is_vertex_cover(G, C):
    for start in (G.get_graph()).keys():   
        for end in G.get_graph()[start]:  
            if not(start in C or end in C):
                return False
    return True
```

### Part 2.1

```python
def present_degree(graph):
    degrees = []
    keys = []
    for key in vertex(graph):
            degree = len(graph.get_edges(key))
            keys.append(key)
            degrees.append(degree)
            
    return (degrees,keys)

def Approx1(graph):
        g_copy = copy.deepcopy(graph)
        covering_vs = set()
                                    
        while True:
            degrees, keys = present_degree(g_copy)
            index = degrees.index(max(degrees))
            high_d_v = keys[index]
            covering_vs.add(high_d_v)
            
            if is_vertex_cover(graph, covering_vs):
                return covering_vs
            
            g_copy.remove_edge(high_d_v)
```

### Part 2.2

```python
def Approx2(graph):
    C = set()
    vertexs = vertex(graph)
    
    while len(vertexs) != 0:
        v = random.choice(vertexs)
        vertexs = [n for n in vertexs if n != v] # Removing v so that we dont get the same vertex again
        C.add(v)
        
        if is_vertex_cover(graph, C):
            return C    
```

### Part 2.3

```python
def Approx3(graph):
    g = copy.deepcopy(graph)
    covering_vs = set()

    while True:
        edges = g.get_all_edges()
        rand_edge = random.choice(edges)

        u = rand_edge[0]
        v = rand_edge[1]
        covering_vs.add(u)
        covering_vs.add(v)
        
        g.remove_edge(u)
        g.remove_edge(v)
        
        if is_vertex_cover(graph, covering_vs):
            return covering_vs        
            
    return covering_vs
```

### Part 2.4

```python
def add_to_each(sets, element):
    copy = sets.copy()
    for set in copy:
        set.append(element)
    return copy

def power_set(set):
    if set == []:
        return [[]]
    return power_set(set[1:]) + add_to_each(power_set(set[1:]), set[0])

            
def MVC(G):
    nodes = list(G.get_nodes())
    subsets = power_set(nodes)
    min_cover = nodes
    for subset in subsets:
        if is_vertex_cover(G, subset):
            if len(subset) < len(min_cover):
                min_cover = subset
    return min_cover
```

```python
# Initialize variables
runs = 100
no_nodes = 6
no_edges1 = 1
sum_min_vc = []
sum_approx1_vc = []
sum_approx2_vc = []
sum_approx3_vc = []

# Run the experiment
for index in range(runs):
    my_graph1 = create_random_graph(no_nodes, no_edges1)
    
    min_vc = len(MVC(my_graph1))
    sum_min_vc.append(sum_min_vc[-1] + min_vc if sum_min_vc else min_vc)

    approx1_vc = len(Approx1(my_graph1))
    sum_approx1_vc.append(sum_approx1_vc[-1] + approx1_vc if sum_approx1_vc else approx1_vc)

    approx2_vc = len(Approx2(my_graph1))
    sum_approx2_vc.append(sum_approx2_vc[-1] + approx2_vc if sum_approx2_vc else approx2_vc)

    approx3_vc = len(Approx3(my_graph1))
    sum_approx3_vc.append(sum_approx3_vc[-1] + approx3_vc if sum_approx3_vc else approx3_vc)

# Print the cumulative sums
print("Minimum Vertex Cover:", sum_min_vc)
print("Approximation 1:", sum_approx1_vc)
print("Approximation 2:", sum_approx2_vc)
print("Approximation 3:", sum_approx3_vc)

plt.plot(sum_min_vc, label='Minimum Vertex Cover')
plt.plot(sum_approx1_vc, label='Approximation 1')
plt.plot(sum_approx2_vc, label='Approximation 2')
plt.plot(sum_approx3_vc, label='Approximation 3')

plt.xlabel('Runs')
plt.ylabel('Sum of Vertex Covers')
plt.title('Comparison of Approximation Methods')
plt.legend()
plt.show()
```

### Reflection on the Experiment Results (Number of edges = 5):

#### When the number of edges is set to 5 and the number of nodes to 6, we can observe two potential types of graphs: <br>
- **Linearly Connected Graphs:** In the first scenario, each node in the graph is connected to at least one other node, resulting in a linearly connected graph. Here, the first and last nodes possess a degree of 1, as they are linked to only one other node, while the intermediary nodes have degrees of 2, indicating connections with two other nodes. This structure ensures the absence of isolated nodes within the graph.
- **Disjointed Graphs:** Alternatively, there exists the possibility of multiple disconnected graphs, with a high chance of forming cyclic patterns.
    

#### Evaluation of Approximation Methods:
- **Approximation 1:**
    - In linearly connected graphs, Approx1 systematically selects vertices with a degree of 2, typically found in the intermediary nodes, which are crucial components of the minimum vertex cover. Thus, it often returns the minimum vertex cover efficiently.
    - In disjointed graphs or when the graph is cyclic, Approx1 still returns the minimum vertex cover. Due to it's nature of implementation contributing to an efficient cover selection.
    - Consequently, the sum increase is similar to the MVC.
- **Approximation 2:**
    - Approx2 selects random vertices across the graph, introducing some diversity in the selected vertices. Although less effective than Approx1, it still outperforms Approx3 due to this strategy.
    - However, Approx2 may fail to return the MVC as the randomly selected vertex could be an isolated vertex (When Disjoint Graph) or one of the nodes with a degree of 1 in linearly connected graphs, neither of which are necessarily part of the vertex cover.
- **Approximation 3:**
    - Approx3 performs the least effectively among the three methods when the number of edges is 5. It adds both vertices connected by a random edge to the vertex cover, regardless of their degrees, leading to suboptimal results, especially in linearly connected graphs where vertices with a degree of 1 are included in the cover, increasing the sum unnecessarily.

#### Evaluation of Approximation Methods:
- **Approximation 1:**
    - Approximation 1 often yields results similar to the minimum vertex cover because it systematically selects nodes with a higher degree, which are more likely to be part of the minimum cover. In this case, as all nodes have the same degree, hence the approximation returns the minimum vertex cover accurately.
- **Approximation 2:**
    - Similarly, Approximation 2 often produces results similar to the minimum vertex cover due to its random selection of vertices. With all nodes having the same degree, the random selection process has an equal chance of choosing nodes that are part of the minimum cover.
- **Approximation 3:**
    - Unlike the other methods, Approximation 3 does not consistently return the minimum vertex cover. This is because it adds both nodes connected by an edge to the cover, leading to redundancy. In this case the approximation returns all the six nodes in the minimum vertex cover hence having a larger sum

### Conclusion :

In the context of our experiment, Approx1 emerges as the most reliable approximation method. Its strategy of selecting vertices with the highest degree ensures that the resulting vertex cover closely resembles the minimum vertex cover (MVC). This is because high-degree vertices tend to have a greater influence on covering edges, making them more likely candidates for inclusion in the MVC. Thus, Approx1 consistently produces vertex covers that align closely with the MVC.

On the other hand, Approx2 and Approx3 exhibit less favorable performance compared to Approx1. Both methods rely on selecting random vertices, which introduces a degree of unpredictability into their outcomes. In many instances, these randomly selected vertices may not contribute significantly to covering edges and may not even be part of the MVC. Consequently, both Approx2 and Approx3 often fail to return the MVC, as their selections may overlook critical vertices necessary for an optimal cover.

Moreover, Approx3 performs worse than Approx2 due to its unique approach of including both vertices connected by a randomly selected edge in the vertex cover. While this may seem intuitive at first, it often leads to the inclusion of redundant vertices and unnecessary duplication within the cover. As a result, Approx3 tends to perform even poorer than Approx2, further deviating from the MVC.

In summary, while Approx1 demonstrates consistent and reliable performance by leveraging high-degree vertices, Approx2 and Approx3 falter due to their reliance on random selections and less efficient vertex cover strategies. Understanding these nuances can guide the selection of the most appropriate approximation method based on the specific characteristics and requirements of the problem at hand.

This is attributed to its systematic selection of nodes with higher degrees, which are more likely to be part of the minimum cover. However, as the number of nodes increases, the chances of selecting the exact same nodes as those in the MVC decrease, leading to occasional deviations from the MVC.
- **Approximation 2:**
    - In contrast, Approximation 2 tends to deviate significantly from the MVC as the number of nodes increases. This divergence can be attributed to its random selection strategy, which becomes less effective with a larger number of nodes. As the pool of available nodes grows, the likelihood of selecting nodes that do not contribute to an optimal vertex cover increases, resulting in vertex covers that differ greatly from the MVC.
- **Approximation 3:**
    - Similar to Approximation 2, Approximation 3 also exhibits substantial deviations from the MVC as the number of nodes increases. Despite its variation in implementation, Approximation 3 faces similar challenges as Approximation 2 due to its reliance on random selection. Consequently, the vertex covers produced by Approximation 3 tend to deviate significantly from the MVC.

In summary, while Approximation 1 maintains relatively close alignment with the MVC even as the number of nodes increases, Approximations 2 and 3 experience greater discrepancies due to their reliance on random selection strategies, which become less effective in larger graph sizes.

```python
def present_degree(graph):
    degrees = []
    keys = []
    for key in vertex(graph):
            degree = len(graph.get_edges(key))
            keys.append(key)
            degrees.append(degree)
            
    return (degrees,keys)

def is_indep_set(graph, ind_set: set):
        if len(ind_set) == 0: return True
        edges = graph.get_all_edges()
        for v in ind_set:
            for u in ind_set:
                edge = [u, v]
                if edge in edges:
                    return False
        return True
    
def find_ind_set(graph):
        ind_vs = set(graph.get_nodes())
        degrees, keys = present_degree(graph)
        
        while not is_indep_set(graph, ind_vs):
            index = degrees.index(max(degrees))
            degrees.pop(index)
            ind_vs.remove(keys[index])
            keys.pop(index)
        return ind_vs
```

```python
def draw_2_graphs(y1, y2, title):
    if not (len(y1) == len(y2)):
        raise ValueError("y1 and y2 should have same length for plotting")
    x_values = [i for i in range(len(y1))]
    plt.figure()
    plt.plot(x_values, y1, label='mvc_scale', marker='o')
    plt.plot(x_values, y2, label='mis_scale', marker='s')
    plt.xlabel('number of edges')
    plt.ylabel('number of elements in set')
    plt.title(title)
    plt.legend()
    plt.show()
```

```python
def add_to_each(sets, element):
    copy = sets.copy()
    for set in copy:
        set.append(element)
    return copy

def power_set(set):
    if set == []:
        return [[]]
    return power_set(set[1:]) + add_to_each(power_set(set[1:]), set[0])

            
def MVC(G):
    nodes = list(G.get_nodes())
    subsets = power_set(nodes)
    min_cover = nodes
    for subset in subsets:
        if is_vertex_cover(G, subset):
            if len(subset) < len(min_cover):
                min_cover = subset
    return min_cover
```

```python
def mvc_mis_relation(v_num):
    trial = 50
    mvc_scales = [0 for i in range(trial)]
    mis_scales = [0 for i in range(trial)]
    
    for i in range(trial):
        e_num = i * 2 + 1
        g_obj = create_random_graph(v_num, e_num)
        
        len_mis = find_ind_set(g_obj)
        mis_scales[i] = len(len_mis)
        
        len_mvc = MVC(g_obj)
        mvc_scales[i] = len(len_mvc)
    
    return [mvc_scales, mis_scales]

[mvc_scales, mis_scales] = mvc_mis_relation(15)
draw_2_graphs(mvc_scales, mis_scales, "No. Elements v/s No. Edges for MIS & MVC")
```

**Observation:**
   - When you sum the size of the MIS and the size of the MVC in a graph, you'll find that the sum equals the total number of vertices in the graph. This is because the vertices included in the MVC cover all edges, leaving the remaining vertices in the MIS with no adjacent vertices, and vice versa.

**Relationship:**
   - The size of the MVC and the size of the MIS are related by the equation:  
     Total number of vertices = Size of MVC + Size of MIS
   - This relationship demonstrates that as the size of the MVC increases, the size of the MIS decreases, and vice versa. Therefore, in any graph, the size of the MVC and the size of the MIS are complementary.
